## Demo: How to work with database


Download database file.


```python
!wget --no-check-certificate https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
df = pd.read_sql_query('SELECT * FROM master',conn)
df.ConvertedComp.plot(kind='hist', figsize=(10,6));
```

### Box Plots


Plot a box plot of `Age.`


```python
df.Age.plot(kind='box', figsize=(10,6));
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6))
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', alpha=0.7, hue='Age', data=df, sizes=(20,800));
plt.figure(figsize=(10,10));
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query="""
SELECT DatabaseDesireNextYear, COUNT(*) as Count 
FROM DatabaseDesireNextYear 
GROUP BY DatabaseDesireNextYear 
ORDER BY COUNT(DatabaseDesireNextYear) DESC LIMIT 5 
"""
df1 = pd.read_sql_query(query,conn)
df1.set_index('DatabaseDesireNextYear', inplace=True)

colors_list=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df1.Count.plot(kind='pie', autopct='%1.1f%%', pctdistance=1.12, startangle=90, shadow=True, labels=None, colors=colors_list,
               figsize=(22,8))
plt.axis('equal')
plt.legend(df1.index, loc = "upper left")
plt.title('Top 5 Databases Respondents Wish To Learn', y=1.12, fontsize=20) 
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master WHERE Age BETWEEN 30 and 35
"""
df2 = pd.read_sql_query(query,conn)
ages = df2.groupby('Age').median()

ages.plot(kind='bar', stacked=True, figsize=(10,8))
plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for ages between 30 and 35');
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
query = """
SELECT ConvertedComp, Age 
FROM master WHERE Age BETWEEN 45 and 60
"""

df3 = pd.read_sql_query(query,conn)
Ages = df3.groupby('Age').median()

Ages.plot(kind='line', figsize=(15,10), color='red')
plt.title('Median ConvertedComp for ages between 45 and 60')
plt.ylabel('ConvertedComp');
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
query="""
SELECT MainBranch, COUNT(*) as Count 
FROM master GROUP BY MainBranch
"""
df4 = pd.read_sql_query(query,conn)
df4.set_index('MainBranch', inplace=True)
df4.plot(kind='barh', figsize=(15,10), color='green')
plt.title('MainBranch counts', fontsize=20, y=1.02);
```

Close the database connection.


```python
conn.close()
```
